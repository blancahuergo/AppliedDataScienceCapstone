```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

#### This will scrape the data from the Wikipedia page using the Python BeautifulSoup library

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
website = requests.get(url)
data = website.text
```

```python
soup = BeautifulSoup(data, "html.parser")
```

```python
website_table = soup.table
```

```python
table_rows = website_table.find_all("tr")
table_rows[0:5]
```

#### Here we will convert the table in the website to a Pandas DataFrame

```python
result = []
for tr in table_rows:
    td = tr.find_all("td")
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        result.append(row)
toronto_df = pd.DataFrame(result, columns = ["Postcode", "Borough", "Neighborhood"])
```

```python
toronto_df.head()
```

#### We will now wrangle the DataFrame to work with rows with NaN values

```python
toronto_df.replace("Not assigned", np.nan, inplace=True)
toronto_df.head()
```

```python
toronto_df.dropna(axis=0, subset=["Borough"], inplace=True)
toronto_df
```

```python
toronto_df["Neighborhood"].replace(np.nan, toronto_df["Borough"], inplace=True)
toronto_df
```

#### We will now group the neighborhoods with the same postal code together

```python
toronto_df = toronto_df.groupby(["Postcode", "Borough"])["Neighborhood"].apply(lambda x: ", ".join(x)).reset_index()
toronto_df
```
