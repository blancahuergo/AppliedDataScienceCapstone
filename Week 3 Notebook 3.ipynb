```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

#### This will scrape the data from the Wikipedia page using the Python BeautifulSoup library

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
website = requests.get(url)
data = website.text
```

```python
soup = BeautifulSoup(data, "html.parser")
```

```python
website_table = soup.table
```

```python
table_rows = website_table.find_all("tr")
table_rows[0:5]
```

#### Here we will convert the table in the website to a Pandas DataFrame

```python
result = []
for tr in table_rows:
    td = tr.find_all("td")
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        result.append(row)
toronto_df = pd.DataFrame(result, columns = ["Postcode", "Borough", "Neighborhood"])
```

```python
toronto_df.head()
```

#### We will now wrangle the DataFrame to work with rows with NaN values

```python
toronto_df.replace("Not assigned", np.nan, inplace=True)
toronto_df.head()
```

```python
toronto_df.dropna(axis=0, subset=["Borough"], inplace=True)
toronto_df
```

```python
toronto_df["Neighborhood"].replace(np.nan, toronto_df["Borough"], inplace=True)
toronto_df
```

#### We will now group the neighborhoods with the same postal code together

```python
toronto_df = toronto_df.groupby(["Postcode", "Borough"])["Neighborhood"].apply(lambda x: ", ".join(x)).reset_index()
toronto_df
```

```python
toronto_df.shape
```

#### This function from the Foursquare lab will save me future time when extracting data

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

#### This function from the clustering lab will simplify the process of extracting data from each neighborhood

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    LIMIT=100
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        results = requests.get(url).json()["response"]['groups'][0]['items']
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood',
                  'Neighborhood Latitude',
                  'Neighborhood Longitude',
                  'Venue',
                  'Venue Latitude',
                  'Venue Longitude',
                  'Venue Category']
    
    return(nearby_venues)
```

#### I will now create a new DataFrame storing the venues using one-hot encoding. This means each row will have a 1 at the intersection of the neighborhood the venue is found in (row) and the type of venue it is (column).

```python
toronto_onehot = pd.get_dummies(toronto_venues[["Venue Category"]], prefix="", prefix_sep="")
toronto_onehot["Neighborhood"] = toronto_venues["Neighborhood"]
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]
```

```python
toronto_onehot.head()
```

#### I will now group these venues into the neighborhood they are in and then convert the values to values between 0 and 1 corresponding to the frequency of that venue in the neighborhood.

```python
toronto_grouped = toronto_onehot.groupby("Neighborhood").mean().reset_index()
```

```python
toronto_grouped.head()
```

#### To explore the newly-extracted data, I will print each neighborhood and the top 10 most frequent types of venue

```python
num_top_venues = 10

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

#### We can store this in a DataFrame showing the top 10 types of venue for each neighborhood

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

#### After the exploratory data analysis, I am ready to use K-Means clustering on the Toronto neighborhoods dataset

```python
toronto_grouped_clustering = toronto_grouped.drop("Neighborhood", axis=1)
toronto_grouped_clustering.head()
```

```python
kclusters = 3
kmeans = KMeans(init="k-means++", n_clusters = kclusters, random_state=0).fit(toronto_grouped_clustering)
```

#### After having fitted the k-means algorithm, let's observe the clusters for the first 15 neighborhoods

```python
kmeans.labels_[:15]
```

#### I will now add a column with the cluster it belongs to for every neighborhood

```python
neighborhoods_venues_sorted.insert(0, "Cluster", kmeans.labels_)
neighborhoods_venues_sorted.head()
```

#### I will now merge this table with the table containing the data of the location and borough of each neighborhood to have a DataFrame containing all the information

```python
toronto_merged = toronto_df3
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index("Neighborhood"), on="Neighborhood")
toronto_merged.head()
```

#### Knowing what neighborhood belongs to each cluster, I can now visualize this on a map

```python
map_clusters = folium.Map(location = toronto_coordinates, zoom_start=11)
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
for lat, lon, neigh, cluster in zip(toronto_merged["Latitude"], toronto_merged["Longitude"], toronto_merged["Neighborhood"], toronto_merged["Cluster"]):
    label = folium.Popup(str(neigh) + " Cluster " + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
```
